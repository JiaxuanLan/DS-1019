```python
import pandas as pd
import numpy as np
```

```python
df_2020 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2020.csv')
df_2020.head()
```

```python
df_2019 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2019.csv')
df_2019.head()
```

```python
df_2020['Issue Date']= pd.to_datetime(df_2020['Issue Date']) 
df_2019['Issue Date']= pd.to_datetime(df_2019['Issue Date']) 
```

```python
df_2020['Year'] = df_2020['Issue Date'].dt.year 
df_2020['Month'] = df_2020['Issue Date'].dt.month
df_2020['Day'] = df_2020['Issue Date'].dt.day
df_2019['Year'] = df_2019['Issue Date'].dt.year 
df_2019['Month'] = df_2019['Issue Date'].dt.month
df_2019['Day'] = df_2019['Issue Date'].dt.day
```

```python
df_2019_fa = df_2020[df_2020['Year'] == 2019]
df_2019_fa
```

```python
df_2019_sp = df_2019[df_2019['Year'] == 2019]
df_2019_sp
```

```python
df_2019_total = pd.concat([df_2019_sp, df_2019_fa])
```

```python
df_2019_total.to_csv('df_2019_total.csv', index = False)
```
