```python
def get_location(df):
    temp_df = pd.DataFrame(df['Street Name'].value_counts())

    temp_df = temp_df[temp_df > 100].dropna().reset_index().rename({'Street Name': 'count', 'index': 'Street Name'}, axis=1)
    temp_df['Street Name'].tolist()
    street_name = temp_df.iloc[:, 0].tolist()
    len(street_name)
    
    Latitude = []
    Longitude = []
    new_street_name = street_name
    remove_list = []

    number = len(street_name)
    #number = 40

    for i in range(number):
        address = street_name[i] + ', NY'
        url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'

        response = requests.get(url).json()
        #print(i)
        if len(response) == 0:
            remove_list.append(i)
            continue
        Latitude.append(float(response[0]["lat"]))
        Longitude.append(float(response[0]["lon"]))
        
    df_position = pd.DataFrame(Latitude,Longitude).reset_index().rename({0: 'Latitude', 'index': 'Longitude'}, axis=1)

    
    count_list = df['Street Name'].value_counts().tolist()
    temp_count_list = count_list[:number]

    reversed_remove_list = sorted(remove_list, reverse=True)
    reversed_remove_list
    for i in reversed_remove_list:
        del temp_count_list[i]
        
    df_position['count'] = temp_count_list
    
    return df_position
    
```
