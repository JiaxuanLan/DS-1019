```python
import pandas as pd
import numpy as np
```

```python
df_2020 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2020.csv')
df_2020.head()
```

```python
df_2021 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2021.csv')
df_2021.head()
```

```python
df_2020['Issue Date']= pd.to_datetime(df_2020['Issue Date']) 
df_2021['Issue Date']= pd.to_datetime(df_2021['Issue Date']) 
```

```python
df_2020['Year'] = df_2020['Issue Date'].dt.year 
df_2020['Month'] = df_2020['Issue Date'].dt.month
df_2020['Day'] = df_2020['Issue Date'].dt.day
df_2021['Year'] = df_2021['Issue Date'].dt.year 
df_2021['Month'] = df_2021['Issue Date'].dt.month
df_2021['Day'] = df_2021['Issue Date'].dt.day
```

```python
df_2020_sp = df_2020[df_2020['Year'] == 2020]
df_2020_sp
```

```python
df_2020_fa = df_2021[df_2021['Year'] == 2020]
df_2020_fa
```

```python
df_2020_total = pd.concat([df_2020_sp, df_2020_fa])
```

```python
df_2020_total.to_csv('df_2020_total.csv', index = False)
```
