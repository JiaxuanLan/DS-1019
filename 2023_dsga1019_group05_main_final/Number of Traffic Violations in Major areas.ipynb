```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
```

```python
df_2018 = pd.read_csv('df_position_2018.csv')
df_2018
```

```python
# import street map
street_map = gpd.read_file('geo_export_f287f26b-51ab-433f-8f6a-5ec85917f725.shp')

# designate coordinate system
crs = {'init':'EPSG:4326'}
# zip x and y coordinates into single feature
geometry = [Point(xy) for xy in zip(df_2018['Longitude'], df_2018['Latitude'])]
# create GeoPandas dataframe
geo_df = gpd.GeoDataFrame(df_2018, crs = crs, geometry = geometry)

# create figure and axes, assign to subplot
fig, ax = plt.subplots(figsize=(15,15))
# add .shp mapfile to axes
street_map.plot(ax=ax, alpha=0.4,color='grey')
# add geodataframe to axes
# assign ‘price’ variable to represent coordinates on graph
# add legend
# make datapoints transparent using alpha
# assign size of points using markersize
geo_df.plot(column='count',ax=ax,alpha=0.5, legend=True, markersize=df_2018['count']/500)
# add title to graph
plt.title('Parking Violation of NYC in 2018', fontsize=15,fontweight='bold')
# set latitiude and longitude boundaries for map display
plt.xlim(-74.05,-73.7)
plt.ylim(40.5,40.94)
# show map
plt.show()
```

```python
df_2018_new = df_2018.copy(deep=True)
df_2018_new = df_2018_new[(df_2018_new['count'] > 7500) == True]
df_2018_new
```
