```python
import pandas as pd
import numpy as np
```

```python
df_2018 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2018.csv')
df_2018.head()
```

```python
df_2019 = pd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2019.csv')
df_2019.head()
```

```python
df_2018['Issue Date']= pd.to_datetime(df_2018['Issue Date']) 
df_2019['Issue Date']= pd.to_datetime(df_2019['Issue Date']) 
```

```python
df_2018['Year'] = df_2018['Issue Date'].dt.year 
df_2018['Month'] = df_2018['Issue Date'].dt.month
df_2018['Day'] = df_2018['Issue Date'].dt.day
df_2019['Year'] = df_2019['Issue Date'].dt.year 
df_2019['Month'] = df_2019['Issue Date'].dt.month
df_2019['Day'] = df_2019['Issue Date'].dt.day
```

```python
df_2018_sp = df_2018[df_2018['Year'] == 2018]
df_2018_sp
```

```python
df_2018_fa = df_2019[df_2019['Year'] == 2018]
df_2018_fa
```

```python
df_2018_total = pd.concat([df_2018_sp, df_2018_fa])
```

```python
df_2018_total.to_csv('df_2018_total.csv', index = False)
```
